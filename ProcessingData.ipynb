```python
import pandas as pd
import numpy as np
```

```python
processedBooking = pd.read_csv('D://Master/SSP/Projekat/data/processed/ProcessedBooking.csv')
processedAirbnb = pd.read_csv('D://Master/SSP/Projekat/data/processed/ProcessedAirbnb.csv')
```

```python
mergedData = pd.concat([processedBooking, processedAirbnb], ignore_index=True)
```

```python
mergedData.shape
```

```python
mergedData.head()
```

```python
mergedData.columns
```

```python
mergedData.rename(columns={'NumberOfPeople': 'Number Of People',
                            'NumberOfAdults': 'Number Of Adults',
                            'NumberOfChildren': 'Number Of Children',
                            'Commission %': 'Commission Percentage',
                            'NumberOfAdults': 'Number Of Adults',
                            'NumberOfNights': 'Number Of Nights',
                           }, inplace=True)
```

```python
mergedData.columns
```

```python
mergedData['Number Of Nights'].unique()
```

```python
mergedData.drop_duplicates(subset=['Book Number'], keep='first', inplace=True)
```

```python
mergedData['Booked By'].unique()
```

```python
mergedData['Booked By'] = mergedData['Booked By'].str.replace(',', '')
```

```python
mergedData["Book Number"].unique()
```

```python
mergedData["Children's Age"] = mergedData["Children's Age"].str.replace(',', ';')
```

```python
mapping = {'One-Bedroom Apartment with Balcony and Sea View' : 4,
           'Studio with Patio' : 2,
           'Studio with Patio and Sea View' : 2,
           'One-Bedroom Apartment with Patio and Sea View' : 4}

mergedData['Bed'] = mergedData['Accommodation'].replace(mapping)
```

```python
finalBooking = mergedData[mergedData['Advertiser'] == 'Booking']
finalAirbnb = mergedData[mergedData['Advertiser'] == 'Airbnb']
```
